# Notebook 05: Compound Statements (Control Flow)

Compound statements contain groups of statements and control their execution. This notebook covers conditions, iteration, the loop-`else` clause, and modern Structural Pattern Matching.

Official Reference: [Python Language Reference - Compound Statements](https://docs.python.org/3/reference/compound_stmts.html)

## 1. Conditional Statements & Truth Value Testing

The `if` statement evaluates expressions sequentially until one is found to be truthy. In Python, truth value testing is implicit. Any object can be tested for truth value.

### Falsy Objects
By default, an object is considered truthy unless its class defines a `__bool__()` method that returns `False` or a `__len__()` method that returns `0`. Built-in falsy values include:
*   `None`, `False`
*   Zero of any numeric type: `0`, `0.0`, `0j`
*   Empty sequences and collections: `""`, `()`, `[]`, `{}`, `set()`

```python
# Testing truth value implicitly of multiple different falsy types
print("Implicit Falsy Testing:")
falsy_values = [None, False, 0, 0.0, "", (), [], {}, set()]
for val in falsy_values:
    if not val:
        print(f"Value {repr(val)} is implicitly Falsy")
```

## 2. Iteration & The Loop-Else Clause

Python loops (`for` and `while`) feature a unique, often misunderstood control flow mechanism: the **`else` clause**.

### How Loop-Else Works
*   The `else` block executes **only if the loop completes normally** (meaning it iterates through the entire sequence or the condition becomes false).
*   If the loop is terminated prematurely via a **`break` statement**, the `else` block is **skipped**.

This allows you to write search loops without using temporary boolean flags.

```python
# Searching using loop-else
def find_item(target, collection):
    for item in collection:
        if item == target:
            print(f"Found target: {target}!")
            break
    else:
        # This runs ONLY if break was never executed
        print(f"Target {target} not found in collection.")

find_item(3, [1, 2, 3])
find_item(5, [1, 2, 3])
```

## 3. Structural Pattern Matching (PEP 634)

Introduced in Python 3.10, structural pattern matching (`match/case`) provides a powerful way to match patterns, destructure data, and perform conditional logic.

Key features:
*   **Literal Patterns:** Direct matching of values (e.g. `case 200:`).
*   **Sequence Patterns:** Match and unpack list/tuple layouts (e.g. `case [first, *rest]:`).
*   **Mapping Patterns:** Match dictionary structures and keys (e.g. `case {"role": role}:`).
*   **Guards:** Append an `if` expression to a case for fine-grained checks (e.g. `case x if x > 0:`).
*   **Wildcard (`_`):** Matches anything, acting as the fallback `else` case.

### Sequence Syntax with Trailing Parameters
When matching sequences, you can combine a starred wildcard capture with fixed trailing parameters. 
-   **Syntax:** `case (prefix, *middle, suffix):` or `case ("publish", *topics, priority):`.
-   **Mechanics:** This matches sequences that have a specific starting element (e.g., the string `"publish"`), followed by zero or more elements captured as a list into `topics`, and a final trailing element captured into `priority`. The pattern matching engine handles length validation and slice-binding dynamically, matching only if the sequence satisfies all literal conditions and length constraints.

```python
# Pattern matching example
def parse_event(event):
    match event:
        # Literal matching
        case "shutdown":
            return "Shutdown command received"
        # Mapping pattern with keys
        case {"type": "click", "button": btn}:
            return f"Click event on button: {btn}"
        # Guards
        case {"type": "keypress", "key": key} if len(key) == 1:
            return f"Key pressed: {key}"
        # Sequence pattern with starred unpacking
        case ["system", "error", *details]:
            return f"System error sequence with details: {details}"
        # Sequence pattern with trailing parameters (starred followed by fixed priority)
        case ("publish", *topics, priority):
            return f"Publish event to {topics} with priority {priority}"
        # Wildcard fallback
        case _:
            return "Unknown event structure"

print(parse_event("shutdown"))
print(parse_event({"type": "click", "button": "left"}))
print(parse_event(["system", "error", "CPU temp high", "fan failure"]))
print(parse_event(("publish", "news", "sports", "high")))
```
