# Table of Contents: Bottom-Up Python Tutorial

Welcome to the Table of Contents for the Bottom-Up Python Tutorial series. Use the links below to navigate through the modules. Each notebook is runnable and includes theoretical explanations based on the official Python Language Reference, interactive code cells, and hands-on coding challenges.

---

## 📚 Curriculum Outline

### Phase 1: Foundations & Core Logic
* **[Notebook 01: Lexical Structure & Basic Execution](01_lexical_structure.ipynb)**
  * Physical vs. logical lines, indentation parsing (`INDENT`/`DEDENT` tokens), identifiers, keywords, Unicode name rules, and raw/byte/f-string literals.
* **[Notebook 02: The Python Data Model (Fundamentals)](02_data_model.ipynb)**
  * Object identity, types, and values; mutable vs. immutable objects; CPython memory interning (small integers and short strings).
* **[Notebook 03: Expressions and Operations](03_expressions.ipynb)**
  * Operator precedence, comparison chaining efficiency, short-circuit evaluation rules, and bitwise arithmetic (two's complement).
* **[Notebook 04: Simple Statements](04_simple_statements.ipynb)**
  * Evaluation order, PEP 3132 starred unpacking in assignments, augmented assignments (`+=` mutability difference), and namespace deletions (`del`).
* **[Notebook 05: Compound Statements (Control Flow)](05_compound_statements.ipynb)**
  * Conditional `if-elif-else`, loop `else` clauses, loop control, and PEP 634 Structural Pattern Matching (guards, wildcards, sequence/mapping patterns).

### Phase 2: Structures, Functions & OOP
* **[Notebook 06: Built-in Data Structures](06_data_structures.ipynb)**
  * Lists (CPython dynamic overallocation), Tuples (recycling pools), Dictionaries (dense compact hash tables per PEP 468), Set algebra, and list comprehension namespace isolation.
* **[Notebook 07: Functions and Scoping](07_functions_and_scoping.ipynb)**
  * LEGB scoping rules, cell objects for closures, the mutable default arguments gotcha, and strict parameters (PEP 570 positional-only `/` and PEP 3102 keyword-only `*`).
* **[Notebook 08: Exception Handling](08_exception_handling.ipynb)**
  * Exception hierarchy (BaseException vs Exception), chaining (`raise from`), `try-except-else-finally` logic, and PEP 654 Exception Groups.
* **[Notebook 09: Classes and Object-Oriented Programming](09_classes_and_oop.ipynb)**
  * Namespaces, instantiation cycles (`__new__` vs `__init__`), cooperative multiple inheritance with C3 Linearization (MRO), `super()` dispatching, properties, and static/class methods.
* **[Notebook 10: Advanced Data Model (Magic Methods)](10_advanced_data_model.ipynb)**
  * Operator overloading, representations (`__repr__` vs `__str__`), custom attribute access interception (`__getattr__`), and container emulation protocols (`__getitem__`, `__len__`).

### Phase 3: Advanced Abstractions & Tooling
* **[Notebook 11: Type Hinting & Static Analysis](11_type_hinting.ipynb)**
  * Basic annotations, PEP 585 collections generics, PEP 604 union pipes, Generic classes (`Generic[T]`), and structural duck-typing via `Protocol`.
* **[Notebook 12: Iterators and Generators](12_iterators_and_generators.ipynb)**
  * Iterator protocol (`__iter__`/`__next__`), generator execution flow, delegating (`yield from`), and generator methods (`send`, `throw`, `close`).
* **[Notebook 13: Context Managers and Decorators](13_context_managers_decorators.ipynb)**
  * Context manager protocol (`__enter__`/`__exit__`), exit exception suppression, `@contextmanager`, decorator mechanics, and parameterized decorators with `functools.wraps`.
* **[Notebook 14: Metaprogramming & Custom Descriptors](14_metaprogramming.ipynb)**
  * Descriptor protocol (`__get__`/`__set__`), class creation hooks (`__init_subclass__`), custom metaclasses, and Abstract Base Classes (`abc.ABC`).

### Phase 4: Systems, Testing & Internals
* **[Notebook 15: Concurrency & Asynchronous Programming](15_concurrency_and_async.ipynb)**
  * Threading vs. Multiprocessing vs. Async I/O, Global Interpreter Lock (GIL) limitations, and `asyncio` event-loop cooperative multitasking.
* **[Notebook 16: Modules and Packages](16_modules_and_packages.ipynb)**
  * Import model caching (`sys.modules`), path resolution (`sys.path`), relative/absolute imports, executable packages (`__main__.py`), and resolving circular imports.
* **[Notebook 17: Modern Packaging, Dependencies & Build Systems](17_modern_packaging.ipynb)**
  * Modern `pyproject.toml` configurations (PEP 517/518/621), package formats (wheels vs sdists), and dependency builds via `uv`.
* **[Notebook 18: Testing, Mocking & Code Quality](18_testing_and_code_quality.ipynb)**
  * Assertions and fixtures in `pytest`, API mock patching with `unittest.mock`, and formatting rules using `ruff`.
* **[Notebook 19: Memory Management, Weak References & Garbage Collection](19_memory_management.ipynb)**
  * CPython reference counting, generational cyclic garbage collection (`gc` thresholds), weak references (`weakref.ref`), and profiling with `tracemalloc` and `cProfile`.
