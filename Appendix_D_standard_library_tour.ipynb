# Appendix D: Brief Tour of the Standard Library

Python is famous for its **"batteries included"** philosophy. The standard library provides a rich set of modules that handle common programming tasks natively. This appendix provides a quick, practical tour of key utility modules covering system interfaces, file operations, text formatting, math, data compression, and quality control.

Reference: [Brief Tour of the Standard Library](https://docs.python.org/3/tutorial/stdlib.html)


## 1. Operating System Interfaces & File Operations

### The `os` Module
The `os` module provides standard interfaces to interact with the host operating system:
- **`os.getcwd()`**: Returns the current working directory path.
- **`os.chdir(path)`**: Changes the current working directory to `path`.
- **`os.system(command)`**: Runs a shell command on the host terminal (returns the exit code). Note: For modern scripting, `subprocess` is preferred for security and parameter escaping.

### The `shutil` Module
For high-level file operations (copying, moving, and archiving files/directories), Python provides `shutil`:
- **`shutil.copyfile(src, dst)`**: Copies the contents of `src` to `dst` in an optimized binary stream.
- **`shutil.move(src, dst)`**: Recursively moves a file or directory to another location (equivalent to renaming or copying/deleting under the hood).

```python
import os
import shutil

# 1. os module directory utilities
current_dir = os.getcwd()
print(f"Current working directory: {current_dir}")

# 2. High-level file copying using shutil
# Let's write a small temporary text file
with open("temp_source.txt", "w") as f:
    f.write("This is standard library temporary file content.")

# Copy file using shutil
shutil.copyfile("temp_source.txt", "temp_copied.txt")
print("File copied successfully. Copy exists?", os.path.exists("temp_copied.txt"))

# Cleanup
if os.path.exists("temp_source.txt"): os.remove("temp_source.txt")
if os.path.exists("temp_copied.txt"): os.remove("temp_copied.txt")
```

## 2. Text Wrapping, Inspection & Formatting

When formatting output messages, terminal layouts, or log statements, Python provides standard tools to shape text strings:

-   **`pprint` (Pretty Printer):** Formats complex, nested data structures (like deep dictionary or list trees) into human-readable layouts with automatic indent alignments.
-   **`reprlib`:** A variant of `repr()` designed for displaying abbreviated representations of very large or deeply nested containers, truncating excessive entries to prevent terminal flooding.
-   **`textwrap`:** Formats paragraphs of text. Key functions include:
    *   `textwrap.fill(text, width)`: Wraps a single paragraph so that every line is at most `width` characters long.
    *   `textwrap.dedent(text)`: Removes common leading whitespace from every line, which is extremely useful for cleaning up multi-line triple-quoted string constants defined inside indented functions.

```python
import pprint
import reprlib
import textwrap

# 1. pprint vs standard print on nested objects
complex_structure = {
    "user": "Alice",
    "roles": ["admin", "developer"],
    "meta": {"login_count": 42, "ip_addresses": ["192.168.1.1", "10.0.0.1"]}
}
print("Standard print output:")
print(complex_structure)
print("\npprint pretty-print output:")
pprint.pprint(complex_structure, width=40)

# 2. reprlib output truncation
large_list = list(range(100))
print(f"\nStandard repr length: {len(repr(large_list))} chars")
print(f"reprlib abbreviated repr: {reprlib.repr(large_list)}")

# 3. textwrap wrapping & dedenting
indented_raw_text = """
    This is a multi-line string
    with a common indentation of four spaces.
    We will dedent it and then wrap it at 30 characters.
    """
dedented = textwrap.dedent(indented_raw_text).strip()
wrapped = textwrap.fill(dedented, width=35)
print("\nDedented and wrapped output:")
print(wrapped)
```

## 3. Mathematical Utilities & Data Compression

### The `statistics` Module
For basic numerical data operations, `statistics` provides high-speed, standard functions:
- **`statistics.mean(data)`**: Computes the arithmetic mean (average).
- **`statistics.median(data)`**: Finds the middle value of data.
- **`statistics.variance(data)`**: Computes the statistical sample variance.

### The `zlib` Module
For compressing string payloads or binary arrays to save bandwidth or disk space, `zlib` offers direct access to the standard gzip DEFLATE algorithm:
- **`zlib.compress(bytes_data)`**: Returns compressed bytes.
- **`zlib.decompress(compressed_bytes)`**: Restores original byte arrays.

```python
import statistics
import zlib

# 1. Numerical statistics
scores = [85, 90, 78, 92, 88]
print(f"Scores statistics:")
print(f"  Mean: {statistics.mean(scores)}")
print(f"  Median: {statistics.median(scores)}")
print(f"  Variance: {statistics.variance(scores):.2f}")

# 2. zlib data compression ratio demonstration
original_payload = b"Repeat this text string multiple times. " * 20
compressed = zlib.compress(original_payload)
decompressed = zlib.decompress(compressed)

print(f"\nCompression details:")
print(f"  Original size: {len(original_payload)} bytes")
print(f"  Compressed size: {len(compressed)} bytes")
print(f"  Compression Ratio: {len(original_payload) / len(compressed):.1f}x")
print(f"  Restored matches original? {decompressed == original_payload}")
```

## 4. Documentation Quality Control: The `doctest` Module

Python docstrings are not just for humans. The standard **`doctest`** module automatically scans your codebase docstrings, extracts lines that look like interactive Python sessions (prefixed with `>>>`), executes them, and asserts that the outputs match the documented expected results.

This is highly valuable because it ensures your API documentation remains **synchronized** with your actual implementation, preventing documentation rot.

```python
import doctest

def calculate_tax(price: float, tax_rate: float) -> float:
    """
    Computes the total tax amount rounded to 2 decimal places.
    
    Doctest examples:
    >>> calculate_tax(100.0, 0.05)
    5.0
    >>> calculate_tax(19.99, 0.15)
    3.0
    """
    return round(price * tax_rate, 2)

# Run the doctest validation programmatically in this cell
print("Running doctest suite...")
test_results = doctest.testmod(verbose=True)
print(f"\nDoctest results summary: {test_results}")
```
